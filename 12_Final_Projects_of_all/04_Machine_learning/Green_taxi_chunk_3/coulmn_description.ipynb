# Green Taxi Dataset - Column Descriptions

The Green Taxi dataset contains detailed information about trips taken by green taxis in New York City. Below is a description of each column in the dataset:

---

## **1. `vendor_id`**
- **Description**: The ID of the taxi vendor or service provider.
- **Values**:
  - `1`: Creative Mobile Technologies (CMT).
  - `2`: Verifone Inc.

---

## **2. `pickup_datetime`**
- **Description**: The date and time when the trip started.
- **Format**: `YYYY-MM-DD HH:MM:SS`.

---

## **3. `dropoff_datetime`**
- **Description**: The date and time when the trip ended.
- **Format**: `YYYY-MM-DD HH:MM:SS`.

---

## **4. `store_and_fwd_flag`**
- **Description**: Indicates whether the trip record was stored locally and forwarded later.
- **Values**:
  - `Y`: Yes, the trip data was stored locally and forwarded later.
  - `N`: No, the trip data was sent to the database in real-time.

---

## **5. `rate_code`**
- **Description**: The fare rate type applied to the trip.
- **Values**:
  - `1`: Standard rate (default).
  - `2`: JFK Airport flat rate.
  - `3`: Newark Airport.
  - `4`: Nassau or Westchester Counties.
  - `5`: Negotiated fare.
  - `6`: Group ride.

---

## **6. `passenger_count`**
- **Description**: The number of passengers in the taxi.
- **Values**:
  - Numeric values (e.g., `1`, `2`, etc.).
  - Note: The data may not always be accurate due to input errors.

---

## **7. `trip_distance`**
- **Description**: The distance of the trip in miles.
- **Unit**: Miles.

---

## **8. `fare_amount`**
- **Description**: The fare charged for the trip, excluding tips and additional fees.
- **Unit**: USD.

---

## **9. `extra`**
- **Description**: Additional charges, such as:
  - Late-night surcharge.
  - Peak hour surcharge.
- **Unit**: USD.

---

## **10. `mta_tax`**
- **Description**: The mandatory New York State Metropolitan Transportation Authority (MTA) tax.
- **Value**: Fixed at `$0.50` per trip.
- **Unit**: USD.

---

## **11. `tip_amount`**
- **Description**: The tip provided by the passenger (if any).
- **Unit**: USD.

---

## **12. `tolls_amount`**
- **Description**: The amount charged for tolls during the trip (if applicable).
- **Unit**: USD.

---

## **13. `ehail_fee`**
- **Description**: The fee charged for electronic hailing (e-hail) services.
- **Unit**: USD.
- **Note**: This field is often empty or null.

---

## **14. `total_amount`**
- **Description**: The total fare amount, including all charges such as fare, tips, tolls, and taxes.
- **Unit**: USD.

---

## **15. `payment_type`**
- **Description**: The method of payment used for the trip.
- **Values**:
  - `1`: Credit card.
  - `2`: Cash.
  - `3`: No charge.
  - `4`: Dispute.
  - `5`: Unknown.
  - `6`: Voided trip.

---

## **16. `trip_type`**
- **Description**: The type of trip.
- **Values**:
  - `1`: Street-hail.
  - `2`: Dispatch (pre-arranged).

---

## **17. `pickup_latitude`**
- **Description**: The latitude of the pickup location.
- **Unit**: Geographic coordinate.

---

## **18. `pickup_longitude`**
- **Description**: The longitude of the pickup location.
- **Unit**: Geographic coordinate.

---

## **19. `dropoff_latitude`**
- **Description**: The latitude of the dropoff location.
- **Unit**: Geographic coordinate.

---

## **20. `dropoff_longitude`**
- **Description**: The longitude of the dropoff location.
- **Unit**: Geographic coordinate.

---
