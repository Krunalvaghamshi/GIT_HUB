```python

# Check unique values in each column
chequing = ['Mobile Weight','RAM','Front Camera','Back Camera','Battery Capacity','Screen Size','Launched Price (Pakistan)','Launched Price (India)','Launched Price (China)','Launched Price (USA)','Launched Price (Dubai)']
for col in chequing:
    print(f"Number of unique values: {df[col].nunique()}")
    print(f"unique values in {col}")
    print(df[col].unique())
    print("____________________________________")
```

```python


# Clean and convert columns to numeric values
columnas = {
    'Mobile Weight' : 'g',
    'Battery Capacity' : 'mAh',
    'Launched Price (Pakistan)' :  'PKR',
    'Launched Price (India)':  'INR',
    'Launched Price (China)' : 'CNY',
    'Launched Price (Dubai)': 'AED'
    }

# For columns without decimals
for col, unidad in columnas.items():
    if 'USA' in col:
        continue  
    df[f'{col} {unidad}'] = pd.to_numeric(
        df[col].astype(str).str.replace(r'[^\d]', '', regex=True),
        errors='coerce'
    ).fillna(0)

# Special handling for US price which contains decimals
df['Launched Price (USA) USD'] = pd.to_numeric(
    df['Launched Price (USA)'].astype(str).str.replace(r'[^\d.]', '', regex=True),
    errors='coerce'
)
```

```python

# Clean RAM column
def modify_RAM(x):
    x = x.replace("GB", "").strip()
    if "/" in x:
        parts = x.split("/")
        return int(max(map(float, parts)))
    return int(float(x))

df['RAM (GB)'] = df['RAM'].apply(modify_RAM)
```

```python


# Clean Front Camera column
def modify_Front_Camera(x):
    # Remove text in parentheses, slashes, and alphabetic characters
    x=re.sub(r'\(.*?\)|/.*|[A-Za-z]+', '', x)
    # Split by common separators
    parts = re.split(r'\+|,|/', x)
    
    nums = []

    for part in parts:
        try:
            val=float(part.strip())
            nums.append(val)
        except ValueError:
            continue
    if nums:
        return int(sum(nums))  
    else:
        return None    
    
df['Front Camera (MP)'] = df['Front Camera'].apply(modify_Front_Camera)
```

```python

# Clean Back Camera column
def extract_camera_values(x):
    # Remove all non-numeric characters except dots and plus signs
    cleaned = re.sub(r'\(.*?\)|[A-Za-z/]+|[^\d.+]', '', str(x))
    # Split by common separators
    parts = re.split(r'\+|,|\s+', cleaned)
    # Convert to float
    return[float(part) for part in parts if part.strip() != '']

# Apply the function to create a new column
df['Back Camera List'] = df['Back Camera'].apply(extract_camera_values)

# Count the maximum number of cameras
max_camera = df['Back Camera List'].apply(len).max()
# Create new columns for each camera
for i in range(max_camera):
     df[f'Back Camera{i + 1} (MP)'] = df['Back Camera List'].apply(lambda x: x[i] if i< len(x) else None)

```

```python

# Clean Screen Size column
def clean_screen_size(x):
    x = re.sub(r'\(.*?\)|/.*|[A-Za-z]+', '', x)
    try:
        return float(x.strip())
    except ValueError:
        return None
    
df['Screen Size (inches)'] = df['Screen Size'].apply(clean_screen_size)
```

```python


# Drop original columns
df = df.drop(['Mobile Weight','RAM','Front Camera','Back Camera','Battery Capacity','Screen Size','Launched Price (Pakistan)','Launched Price (India)','Launched Price (China)','Launched Price (USA)','Launched Price (Dubai)'],axis=1)
df
```

```python

# Visualization 1: Cellphones quantity per brand (Donut chart)
Cell_phones_by_Company = df.groupby("Company Name")["Model Name"].count().reset_index()
Cell_phones_by_Company.columns = ["Company Name", "Phone Count"]
source = Cell_phones_by_Company
base = alt.Chart(source).encode(
    theta=alt.Theta("Phone Count:Q", stack=True),
     radius=alt.Radius("Phone Count:Q", scale=alt.Scale(type="sqrt", zero=True, rangeMin=20)),
    color="Company Name:N"
)
c1=base.mark_arc(innerRadius=20,stroke="#fff")
c2=base.mark_text(radiusOffset=10).encode(text="Phone Count:Q")
(c1 + c2).properties(
    title="Cellphones quantity per brand",
    width=700,
    height=500
)
```

```python

# Visualization 2: Phones released per brand over the years
cellphones_per_year = df.groupby(['Launched Year','Company Name']).size().reset_index(name='Phone Count')

chart = alt.Chart(cellphones_per_year).mark_bar().encode(
    x=alt.X('Launched Year:O', title='Year'),
    y=alt.Y('Phone Count:Q', title='Phones Released'),
    color=alt.Color('Company Name:N', title="Brand", scale=alt.Scale(scheme='category20')),
    column=alt.Column('Company Name:N', title='Brand'),
    tooltip=['Company Name:N', 'Launched Year:O', 'Phone Count:Q']  
).properties(
    width=200,
    height=200,
    title='Phones Released Per Brand Over the Years'
)

chart
```

```python


# Visualization 3: Launch price trends by brand
cellphones_prices_USA = df.groupby(["Company Name","Launched Year"])["Launched Price (USA) USD"].median().reset_index()
brands = cellphones_prices_USA['Company Name'].unique()
charts = []

for brand in brands:
    brand_data = cellphones_prices_USA[cellphones_prices_USA['Company Name'] == brand] 

    chart = alt.Chart(brand_data).mark_line(point=True).encode(
        x=alt.X("Launched Year:T", title="Release year"),
        y=alt.Y("Launched Price (USA) USD:Q", title="Price (USD)"),
        tooltip=['Company Name:N', 'Launched Year:O', 'Launched Price (USA) USD:Q']
    ).properties(
        title=f"Launch Price per Year (US) - {brand}",
        width=400,
        height=300
    )
    charts.append(chart)

n_cols = 5
rows = [alt.hconcat(*charts[i:i + n_cols]) for i in range(0, len(charts), n_cols)]
mosaic = alt.vconcat(*rows)

mosaic
```
